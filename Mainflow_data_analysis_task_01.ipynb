**BASIC LIST OPERATIONS**

1. Adding element to the list
2. creating a copy of the list
3. returning number of elements with the specified value
4. adding list elements to the end of the current list
5. returning the index of the specified element
6. inserting element at the specified position
7. returning the removed element using pop
8. removing the element with the specified value
9. reversing the list
10. sorting the list
11. removing all the elements from the list

```python
#CREATING A LIST
my_list=[1,2,3,4,5]
print("original list: ",my_list)


#ADDING AN ELEMENT AT THE END OF THE LIST
my_list.append(6)
print("updated list after appending: ",my_list)


#CREATING A COPY OF THE LIST
my_list.copy()
print("updated list after copying: ",my_list)


#RETURNING NUMBER OF ELEMENTS WITH THE SPECIFIED VALUE
print("count of the value 2: ",my_list.count(2))


#ADDING THE ELEMENTS OF A LIST TO THE END OF THE CURRENT LIST
my_list.extend([7,8,9,10])
print("updated list after extending: ",my_list)


#RETURNING THE INDEX OF THE SPECIFIED ELEMENT
print("the index of the element 4 is ",my_list.index(4))


#ADDING AN ELEMENT AT THE SPECIFIED POSITION
my_list.insert(3,11)
print("updated list after inserting 11: ",my_list)


#REMOVING THE ELEMENT AT THE SPECIFIED POSITION
my_list.pop(5)
print("updated list after popping element 5: ",my_list)


#REMOVING THE ITEM WITH THE SPECIFIED VALUE
my_list.remove(1)
print("updated list after removing element 1: ",my_list)


#REVERSING THE LIST
my_list.reverse()
print("updated list after reversing: ",my_list)


#SORTING THE LIST
my_list.sort()
print("updated list after sorting: ",my_list)


#REMOVING ALL THE ELEMENTS FROM THE LIST
my_list.clear()
print("updated list after clearing: ",my_list)
```

**BASIC DICTIONARY OPERATIONS**

1. returning a copy of the dictionary
2. returning a dictionary with the specified keys and values
3. returning the value of the specified key
4. Returning a list containing a tuple for each key value pair
5. Returning a list containing the dictionary's keys
6. Removing the element with the specified key
7. Removing the last inserted key-value pair
8. Returning the value of the specified key. If the key does not exist: insert the key, with the specified value
9. Updating the dictionary with the specified key-value pairs
10.Returning a list of all the values in the dictionary

```python
#CREATING A DICTIONARY
my_dict={'name':'sweetha','age':20,'gender':'female','city':'chennai'}
print("original dictionary:",my_dict)


#CREATING A COPY OF THE DICTIONARY
my_dict.copy()
print("updated dictionary after copying: ",my_dict)


#RETURNING A DICTIONARY WITH THE SPECIFIED KEYS AND VALUES
a=('key1','key2','key3')
b=('value1','value2','value3')
new_dict=my_dict.fromkeys(a,b)
print(new_dict)


#RETURNING THE VALUE OF THE SPECIFIED KEY
print("value of the key 'name' is: ",my_dict.get('name'))


#RETURNING A LIST CONTAINING A TUPLE FOR EACH KEY-VALUE PAIR
print("list of tuples: ",my_dict.items())


#RETURNING A LIST CONTAINING DICTIONARY'S KEYS
print("list of keys: ",my_dict.keys())


#REMOVING THE ELEMENT WITH THE SPECIFIED KEY
my_dict.pop('city')
print("updated dictionary after popping: ",my_dict)


#REMOVING THE LAST INSERTED KEY-VALUE PAIR
my_dict.popitem()
print("updated dictionary after popping item: ",my_dict)


#RETURNING THE VALUE OF THE SPECIFIED KEY. IF THE KEY DOES NOT EXIST: INSERT THE KEY, WITH THE SPECIFIED VALUE
my_dict.setdefault('age',21)
print("updated dictionary after setting default: ",my_dict)


#UPDATING THE DICTIONARY WITH THE SPECIFIED KEY-VALUE PAIRS
my_dict.update({'city':'mumbai'})
print("updated dictionary after updating: ",my_dict)


#RETURNING A LIST OF ALL THE VALUES IN THE DICTIONARY
print("list of values: ",my_dict.values())


#REMOVING ALL ELEMENTS FROM THE DICTIONARY
my_dict.clear()
print("updated dictionary after clearing: ",my_dict)
```

**BASIC SET OPERATIONS**

1. adding an element to the set
2. Removing all the elements from the set
3. Returning a copy of the set
4. Returning a set containing the difference between two or more sets
5. Removing the items in this set that are also included in another, specified set
6. Removing the specified item
7. Returning a set, that is the intersection of two other sets
8. Removing the items in this set that are not present in other, specified set(s)
9. Returning whether two sets have a intersection or not
10.Returning whether another set contains this set or not
11.Returning whether all items in this set is present in other, specified set(s)
12.Returning whether this set contains another set or not
13.Returning whether all items in other, specified set(s) is present in this set
14.Removing an element from the set
16.Returning a set with the symmetric differences of two sets
17.Inserting the symmetric differences from this set and another
18.Returning a set containing the union of sets
18.Updating the set with the union of this set and others

```python
#CREATING A SET
my_set1={1,2,3,4,5,6,7}
my_set2={6,7,8,9,2}
print("first set: ",my_set1)
print("second set: ",my_set2)


#ADDING AN ELEMENT TO THE SET
my_set1.add(18)
print("updated set after adding: ",my_set1)


#RETURNING A COPY OF THE SET
my_set2.copy()
print("updated set after copying: ",my_set2)


#RETURNING A SET CONTAINING DIFFERENCE BETWEEN TWO OR MORE SETS
print("difference between two sets: ",my_set1.difference(my_set2))


#REMOVING THE ITEMS IN THIS SET THAT ARE ALSO INCLUDED IN ANOTHER, SPECIFIED SET
my_set1.intersection_update(my_set2)
print("updated set after intersection update: ",my_set1)



#REMOVING THE SPECIFIED ITEM
my_set1.discard(2)
print("updated set after discarding: ",my_set1)



#RETURNING A SET THAT IS THE INTERSECTION OF TWO OTHER SETS\
print("intersection of two sets: ",my_set1.intersection(my_set2))


#REMOVING THE ITEMS IN THIS SET THAT ARE NOT PRESENT IN OTHER, SPECIFIED SET(S)
my_set1.symmetric_difference_update(my_set2)
print("updated set after symmetric difference update: ",my_set1)


#RETURNING WHETHER TWO SETS HAVE A INTERSECTION OR NOT
print("is there any intersection between two sets: ",my_set1.isdisjoint(my_set2))


#RETURNING WHETHER ANOTHER SET CONTAINS THIS SET OR NOT
print("is my_set1 a subset of my_set2: ",my_set1.issubset(my_set2))


#RETURNING WHETHER ALL ITEMS IN THIS SET IS PRESENT IN OTHER, SPECIFIED SET(S)
print("is my_set1 a superset of my_set2: ",my_set1.issuperset(my_set2))


#RETURNING WHETHER THIS SET CONTAINS ANOTHER SET OR NOT
print("is my_set1 a set of my_set2: ",my_set1.issuperset(my_set2))


#RETURNING WHETHER ALL ITEMS IN OTHER, SPECIFIED SET(S) IS PRESENT IN THIS SET
print("is my_set1 a superset of my_set2: ",my_set1.issuperset(my_set2))


#REMOVING AN ELEMENT FROM THE SET
my_set1.pop()
print("updated set after removing: ",my_set1)



#RETURNING A SET WITH THE SYMMETRIC DIFFERENCE OF TWO SETS
print("symmetric difference of two sets: ",my_set1.symmetric_difference(my_set2))


#INSERTING THE SYMMETRIC DIFFERENCES FROM THIS SET AND ANOTHER
my_set1.symmetric_difference_update(my_set2)
print("updated set after symmetric difference update: ",my_set1)


#RETURNING A SET CONTAINING THE UNION OF SETS
print("union of two sets: ",my_set1.union(my_set2))


#UPDATING THE SET WITH THE UNION OF THIS SET AND OTHERS
my_set1.update(my_set2)
print("updated set after updating: ",my_set1)


#REMOVING ALL THE ELEMENTS FROM THE SET
my_set1.clear()
my_set2.clear()
print("updated set after clearing: ",my_set1)
print("updated set after clearing: ",my_set2)
```
